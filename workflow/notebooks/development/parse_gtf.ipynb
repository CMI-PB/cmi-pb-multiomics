## Gathering all protein coding genes

```python
data = []
gene_types = set()
with gzip.open(gtf_fn) as f:
    for line in f:
        line = line.decode()
        
        if line.startswith('#'):
            continue
            
        info = line.split('\t')

        # extracting the extra column
        extra = info[-1].split()
        extra = {extra[2 * i]: extra[2 * i + 1] for i in range(int(len(extra)/2))}
        gene_id = extra['gene_id'].replace('"', '').replace(';', '')
        gene_type = extra['gene_type'].replace('"', '').replace(';', '')
        gene_name = extra['gene_name'].replace('"', '').replace(';', '')
        #gene_types.add(gene_type)
        
        if gene_type == 'protein_coding' and info[2] == 'gene':

            coding = [info[0], info[3], info[4], info[6], gene_id, gene_name]
            data.append(coding)
```

```python
final_data = set([tuple(x) for x in data])
final_data = list(final_data)
final_data.sort(key=lambda x: (x[0], int(x[1]), int(x[2])))
```

```python
fn = 'results/refs/gencode/gencode.v38lift37.annotation.protein_coding.bed'
with open(fn, 'w') as f:
    for record in final_data: 
        s = '\t'.join(record) + '\n'
        f.write(s)
```
